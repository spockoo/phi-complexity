## 1. L'Espace Z[φ] — Anneau des Entiers Dorés

L'anneau $\mathbb{Z}[\varphi] = \{a + b\varphi \mid a, b \in \mathbb{Z}\}$ est la structure algébrique fondamentale du Framework φ-Meta.

Chaque métrique est projetée dans cet espace via la **Transmutation Maat** :
$$a = \lfloor \text{val} \times 1000 \rfloor \quad (\text{Mati\`ere})$$
$$b = \lfloor \text{val} \times 1370 \rfloor \quad (\text{R\'esonance } \alpha^{-1})$$
$$\text{Coord}_{Z[\varphi]} = b + (a+b)\varphi$$

Le facteur 1370 vient de $1000 / \alpha \approx 1000 \times 137.035 = 137035$, simplifié à $1370$.

```python
# Visualisation de l'espace Z[φ]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Grille Z[φ] : points a + bφ
a_range = range(-5, 15)
b_range = range(-5, 15)

for a in a_range:
    for b in b_range:
        val = a + b * PHI
        ax1.plot(a, b, "o", color="#FFD700" if val > 0 else "#1E90FF",
                markersize=4, alpha=0.6)

# Axes
ax1.axhline(y=0, color="gray", linewidth=0.5)
ax1.axvline(x=0, color="gray", linewidth=0.5)

# Points remarquables
points_speciaux = [
    (1, 0, "1"),
    (0, 1, "φ"),
    (-1, 1, "φ-1=φ⁻¹"),
    (1, 1, "1+φ=φ²"),
    (2, 1, "2+φ"),
]
for a, b, label in points_speciaux:
    ax1.plot(a, b, "*", color="#FF6347", markersize=12, zorder=5)
    ax1.annotate(label, (a, b), textcoords="offset points", xytext=(8, 8), fontsize=9)

ax1.set_xlabel("a (Composante enti\u00e8re)", fontsize=12)
ax1.set_ylabel("b (Composante φ)", fontsize=12)
ax1.set_title("R\u00e9seau Z[φ] \u2014 Anneau des Entiers Dor\u00e9s", fontsize=13)
ax1.grid(True, alpha=0.2)

# Distribution des valeurs a + bφ
vals = sorted([a + b * PHI for a in range(0, 20) for b in range(0, 20)])
ax2.hist(vals, bins=50, color="#FFD700", edgecolor="black", alpha=0.8)
ax2.set_xlabel("Valeur a + bφ", fontsize=12)
ax2.set_ylabel("Fr\u00e9quence", fontsize=12)
ax2.set_title("Distribution des valeurs dans Z[φ]⁺", fontsize=13)

plt.suptitle("L'Espace Holographique Z[φ]", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 2. Transmutation Maat — Projection des Métriques

Projetons les métriques de fichiers Python réels dans l'espace Z[φ].

```python
# Collecter les métriques de fichiers réels
fichiers = []
for root, dirs, files in os.walk(os.path.join("..", "phi_complexity")):
    for f in files:
        if f.endswith(".py") and not f.startswith("__"):
            fichiers.append(os.path.join(root, f))

metriques_all = []
for fpath in fichiers:
    try:
        m = auditer(fpath)
        m["_basename"] = os.path.basename(fpath)
        metriques_all.append(m)
    except Exception:
        pass

print(f"Fichiers analys\u00e9s : {len(metriques_all)}")

# Transmutation Maat pour chaque fichier
fig, ax = plt.subplots(figsize=(12, 8))

for m in metriques_all:
    mat = MatriceHolographique(m)
    coords = mat.transmuter()
    masse = mat.calculer_masse_harmonique()

    # Projection 2D : premi\u00e8re coordonn\u00e9e Z[φ] vs deuxi\u00e8me
    if len(coords) >= 2:
        x_val = coords[0]["a"] + coords[0]["b"] * PHI
        y_val = coords[1]["a"] + coords[1]["b"] * PHI

        color = "#FFD700" if m["radiance"] >= 85 else "#1E90FF" if m["radiance"] >= 60 else "#FF6347"
        ax.scatter(x_val, y_val, c=color, s=masse * 500 + 50,
                   edgecolors="black", linewidth=0.5, alpha=0.8, zorder=5)
        ax.annotate(m["_basename"][:12], (x_val, y_val),
                   fontsize=7, ha="center", va="bottom",
                   textcoords="offset points", xytext=(0, 8))

ax.set_xlabel("Coordonn\u00e9e Maat (Radiance) \u2014 a + bφ", fontsize=12)
ax.set_ylabel("Coordonn\u00e9e Maat (R\u00e9sistance) \u2014 a + bφ", fontsize=12)
ax.set_title("Transmutation Maat \u2014 Projection Z[φ] des Fichiers\n"
             "(Taille = Masse Harmonique, Or=Herm\u00e9tique, Bleu=\u00c9veil, Rouge=Dormant)",
             fontsize=13)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Dôme de Résonance

Le dôme de résonance est une visualisation multi-axes de la santé harmonique d'un fichier.

```python
# Dôme de résonance pour les 6 premiers fichiers
n_display = min(6, len(metriques_all))
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

for idx in range(n_display):
    ax = axes[idx // 3, idx % 3]
    m = metriques_all[idx]
    mat = MatriceHolographique(m)

    # Radar des 5 axes
    labels_axes = ["Radiance", "R\u00e9sistance", "Lilith\n(inv.)", "Shannon\n(inv.)", "φ-Ratio"]
    values = mat.vecteur.copy()
    # Inverser variance et entropie (plus bas = mieux)
    values[2] = 1.0 - values[2]
    values[3] = 1.0 - values[3]
    values[4] = 1.0 - values[4]

    N = len(labels_axes)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    values_plot = values + [values[0]]
    angles += [angles[0]]

    ax_polar = fig.add_axes(ax.get_position(), projection="polar")
    ax.set_visible(False)

    ax_polar.fill(angles, values_plot, color="#FFD700", alpha=0.3)
    ax_polar.plot(angles, values_plot, color="#FFD700", linewidth=2)
    ax_polar.scatter(angles[:-1], values, color="#FFD700", s=50, zorder=5)

    ax_polar.set_xticks(angles[:-1])
    ax_polar.set_xticklabels(labels_axes, fontsize=8)
    ax_polar.set_ylim(0, 1)
    ax_polar.set_title(f"{m['_basename'][:15]}\nR={m['radiance']:.0f}", fontsize=10, pad=15)

plt.suptitle("D\u00f4mes de R\u00e9sonance \u2014 MatriceHolographique", fontsize=15, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 4. Similitude Cosinus entre Architectures

La similitude cosinus mesure la \u00ab distance structurelle \u00bb entre deux fichiers dans l'espace φ.

```python
# Matrice de similitude cosinus
n_files = min(10, len(metriques_all))
noms = [m["_basename"][:12] for m in metriques_all[:n_files]]

matrices = [MatriceHolographique(m) for m in metriques_all[:n_files]]
sim_matrix = np.zeros((n_files, n_files))

for i in range(n_files):
    for j in range(n_files):
        sim_matrix[i, j] = matrices[i].calculer_similitude(matrices[j])

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(sim_matrix, cmap="magma", vmin=0.5, vmax=1.0)
plt.colorbar(im, label="Similitude cosinus", ax=ax)

ax.set_xticks(range(n_files))
ax.set_yticks(range(n_files))
ax.set_xticklabels(noms, rotation=45, ha="right", fontsize=9)
ax.set_yticklabels(noms, fontsize=9)

# Annoter les valeurs
for i in range(n_files):
    for j in range(n_files):
        ax.text(j, i, f"{sim_matrix[i,j]:.2f}", ha="center", va="center",
                fontsize=7, color="white" if sim_matrix[i,j] < 0.8 else "black")

ax.set_title("Matrice de Similitude Cosinus \u2014 Architectures φ", fontsize=14)
plt.tight_layout()
plt.show()

# Trouver les paires les plus similaires (hors diagonale)
max_sim = 0
pair = ("", "")
for i in range(n_files):
    for j in range(i+1, n_files):
        if sim_matrix[i, j] > max_sim:
            max_sim = sim_matrix[i, j]
            pair = (noms[i], noms[j])
print(f"Paire la plus similaire : {pair[0]} \u2194 {pair[1]} (similitude = {max_sim:.4f})")
```

## 5. Visualisation 3D de l'Espace φ

Projetons les fichiers dans un espace 3D (Radiance × Variance × Entropie).

```python
# Visualisation 3D
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection="3d")

radiancies = [m["radiance"] for m in metriques_all]
variances = [m["lilith_variance"] for m in metriques_all]
entropies = [m["shannon_entropy"] for m in metriques_all]
noms_3d = [m["_basename"][:10] for m in metriques_all]

colors_3d = ["#FFD700" if r >= 85 else "#1E90FF" if r >= 60 else "#FF6347" for r in radiancies]

scatter = ax.scatter(radiancies, variances, entropies,
                     c=radiancies, cmap="magma", s=100,
                     edgecolors="black", linewidth=0.5, alpha=0.8)

for i, nom in enumerate(noms_3d):
    ax.text(radiancies[i], variances[i], entropies[i], f"  {nom}",
            fontsize=7, alpha=0.8)

ax.set_xlabel("Radiance", fontsize=11)
ax.set_ylabel("Variance de Lilith", fontsize=11)
ax.set_zlabel("Entropie de Shannon", fontsize=11)
ax.set_title("Espace Holographique 3D\n(Radiance \u00d7 Lilith \u00d7 Shannon)", fontsize=14)

plt.colorbar(scatter, label="Radiance", ax=ax, shrink=0.6)
plt.tight_layout()
plt.show()
```
