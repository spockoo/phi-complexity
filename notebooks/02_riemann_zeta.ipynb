# ζ Fonction Zeta et Résonance du Code

> *La fonction zêta de Riemann révèle les harmoniques cachées des nombres premiers — la fonction ζ_meta de phi-complexity révèle les harmoniques cachées du code.*

Ce notebook explore le parallèle entre la fonction zêta de Riemann et le `zeta_score` de phi-complexity.

**Prérequis** : `pip install phi-complexity[notebooks]`

```python
import math
import sys
sys.path.insert(0, "..")

import matplotlib.pyplot as plt
import numpy as np

from phi_complexity.core import PHI, PHI_INV, ZETA_PLANCHER
```

## 1. La Fonction Zêta de Riemann

$$\zeta(s) = \sum_{n=1}^{\infty} \frac{1}{n^s}$$

Pour $s > 1$, la série converge. Les **zéros non-triviaux** se situent sur la droite critique $\text{Re}(s) = 1/2$ (hypothèse de Riemann).

```python
# Visualisation de ζ(s) pour s réel > 1
s_values = np.linspace(1.01, 10, 500)

def zeta_real(s, n_terms=1000):
    """Approximation de ζ(s) par sommation partielle."""
    return sum(1.0 / (n ** s) for n in range(1, n_terms + 1))

zeta_values = [zeta_real(s) for s in s_values]

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(s_values, zeta_values, color="#FFD700", linewidth=2, label="ζ(s) — Riemann")
ax.axhline(y=PHI, color="#1E90FF", linestyle="--", linewidth=1.5, label=f"φ = {PHI:.4f}")
ax.axhline(y=1, color="gray", linestyle=":", linewidth=1)
ax.axvline(x=PHI, color="#FF6347", linestyle="--", linewidth=1.5, label=f"s = φ ≈ {PHI:.4f}")

# Marquer ζ(φ)
zeta_phi = zeta_real(PHI)
ax.plot(PHI, zeta_phi, "o", color="#FF6347", markersize=10, zorder=5)
ax.annotate(f"ζ(φ) ≈ {zeta_phi:.4f}", xy=(PHI, zeta_phi),
            xytext=(PHI + 0.5, zeta_phi + 0.3), fontsize=11,
            arrowprops=dict(arrowstyle="->", color="black"))

ax.set_xlabel("s", fontsize=13)
ax.set_ylabel("ζ(s)", fontsize=13)
ax.set_title("Fonction Zêta de Riemann — Le point doré ζ(φ)", fontsize=14)
ax.legend(fontsize=11)
ax.set_ylim(0, 5)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"ζ(φ) ≈ {zeta_phi:.6f}")
print(f"ζ(2) = π²/6 ≈ {math.pi**2/6:.6f}")
```

## 2. ζ_meta — La Fonction Zêta de phi-complexity

Dans phi-complexity, le **zeta_score** est défini comme :

$$\zeta_{\text{meta}}(n) = \min\left(1, \frac{\phi}{n} \sum_{i=1}^{n} \frac{1}{(i+1)^{\varphi}}\right)$$

C'est une série convergente qui mesure la **résonance harmonique globale** du code.
Le score converge vers 1.0 pour un code parfaitement structuré.

```python
# Comparaison : ζ_meta(n) pour différents nombres de fonctions
n_range = range(1, 51)

def zeta_meta(n):
    """ζ_meta = min(1, [Σ 1/(i+1)^φ / n] × φ)"""
    zeta = sum(1.0 / ((i + 1) ** PHI) for i in range(n)) / n
    return min(1.0, zeta * PHI)

scores = [zeta_meta(n) for n in n_range]

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(list(n_range), scores, color="#FFD700", linewidth=2.5, label="ζ_meta(n)")
ax.axhline(y=ZETA_PLANCHER, color="#FF6347", linestyle="--", linewidth=1.5,
           label=f"Plancher ζ = φ⁻² ≈ {ZETA_PLANCHER:.4f}")
ax.axhline(y=PHI_INV, color="#1E90FF", linestyle="--", linewidth=1.5,
           label=f"φ⁻¹ ≈ {PHI_INV:.4f}")
ax.axhline(y=1.0, color="gray", linestyle=":", linewidth=1)

ax.fill_between(list(n_range), ZETA_PLANCHER, 0, alpha=0.1, color="red",
                label="Zone DORMANT")
ax.fill_between(list(n_range), 1.0, ZETA_PLANCHER, alpha=0.05, color="gold")

ax.set_xlabel("Nombre de fonctions (n)", fontsize=13)
ax.set_ylabel("ζ_meta(n)", fontsize=13)
ax.set_title("Score Zeta φ-Meta — Résonance vs Nombre de Fonctions", fontsize=14)
ax.legend(fontsize=10, loc="upper right")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"ζ_meta(1)  = {zeta_meta(1):.6f}")
print(f"ζ_meta(5)  = {zeta_meta(5):.6f}")
print(f"ζ_meta(13) = {zeta_meta(13):.6f}")
print(f"ζ_meta(50) = {zeta_meta(50):.6f}")
```

## 3. Les « Zéros de la Radiance »

Par analogie avec les zéros de Riemann, explorons les points où la radiance bascule entre ordre et chaos — les frontières entre les statuts gnostiques.

```python
# Simulation : radiance en fonction de la variance (Lilith) et de l'entropie
variance_range = np.linspace(0, 500, 200)
entropie_range = np.linspace(0, 5, 200)
V, E = np.meshgrid(variance_range, entropie_range)

# Formule simplifiée de radiance
seuil_lilith = PHI**2 * 100  # ≈ 261.8
seuil_entropie = math.log2(PHI**4)  # ≈ 2.88

R = 100.0 - np.minimum(25, (V / seuil_lilith) * 25) - np.minimum(20, np.maximum(0, E - seuil_entropie) * 5)

fig, ax = plt.subplots(figsize=(12, 8))
contour = ax.contourf(V, E, R, levels=20, cmap="magma")
plt.colorbar(contour, label="Radiance R(C)", ax=ax)

# Frontières gnostiques
ax.contour(V, E, R, levels=[60], colors=["cyan"], linewidths=[2], linestyles=["--"])
ax.contour(V, E, R, levels=[85], colors=["gold"], linewidths=[2], linestyles=["--"])

ax.set_xlabel("Variance de Lilith (σ²_L)", fontsize=13)
ax.set_ylabel("Entropie de Shannon (H_S)", fontsize=13)
ax.set_title("Carte de Radiance — Les « Zéros Gnostiques »\n"
             "Lignes : DORMANT↔EN ÉVEIL (60) | EN ÉVEIL↔HERMÉTIQUE (85)", fontsize=13)
ax.grid(True, alpha=0.2, color="white")
plt.tight_layout()
plt.show()
```
