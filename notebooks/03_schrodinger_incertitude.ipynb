# ħ_φ Incertitude de Heisenberg-Phi

> *Comme en mécanique quantique, il existe un compromis irréductible entre la complexité et la lisibilité d'un code.*

$$\Delta C \cdot \Delta L \geq \frac{\hbar_\varphi}{2} \approx 0.309$$

Ce notebook explore la relation d'incertitude de Heisenberg-Phi (CM-HUP) implémentée dans phi-complexity.

**Prérequis** : `pip install phi-complexity[notebooks]`

```python
import math
import sys
sys.path.insert(0, "..")

import matplotlib.pyplot as plt
import numpy as np

from phi_complexity.core import PHI, PHI_INV, HBAR_PHI
from phi_complexity import auditer
```

## 1. La Relation d'Incertitude

En mécanique quantique, Heisenberg a montré que :
$$\Delta x \cdot \Delta p \geq \frac{\hbar}{2}$$

Par analogie, dans le Framework φ-Meta :
- **ΔC** = Incertitude de complexité = $\sqrt{\sigma^2_L / \sigma^2_{max}}$
- **ΔL** = Incertitude de lisibilité = $\min(1, H_F / H_{max})$
- **ħ_φ** = Constante d'action réduite dorée = $1/\varphi \approx 0.6180$
- **Plancher** = $\hbar_\varphi / 2 \approx 0.309$

La **tension quantique** = $(ΔC \cdot ΔL) / plancher$ :
- **< 1** : État super-cohérent (code focalisé et élégant)
- **≈ 1** : Optimum golden (compromis parfait)
- **> 1** : Zone d'incertitude naturelle

```python
# Visualisation de la relation d'incertitude
plancher = HBAR_PHI / 2  # ≈ 0.309

delta_c = np.linspace(0.01, 1.5, 300)
delta_l_min = plancher / delta_c  # Frontière d'incertitude minimale

fig, ax = plt.subplots(figsize=(10, 8))

# Zone interdite (sous la courbe d'incertitude)
ax.fill_between(delta_c, 0, np.minimum(delta_l_min, 1.5),
                alpha=0.15, color="red", label="Zone interdite (ΔC·ΔL < ħ_φ/2)")

# Courbe d'incertitude minimale
ax.plot(delta_c, delta_l_min, color="#FF6347", linewidth=2.5,
        label=f"ΔC·ΔL = ħ_φ/2 = {plancher:.4f}")

# Zone super-cohérente
ax.fill_between(delta_c[delta_l_min <= 1.5], delta_l_min[delta_l_min <= 1.5], 1.5,
                alpha=0.05, color="gold")

# Points remarquables
# Point golden : ΔC = ΔL = sqrt(plancher) ≈ 0.556
golden_point = math.sqrt(plancher)
ax.plot(golden_point, golden_point, "*", color="#FFD700", markersize=20, zorder=5,
        label=f"Point Doré ({golden_point:.3f}, {golden_point:.3f})")

# Point φ⁻¹
ax.plot(PHI_INV, plancher / PHI_INV, "D", color="#1E90FF", markersize=10, zorder=5,
        label=f"Point φ⁻¹ ({PHI_INV:.3f}, {plancher/PHI_INV:.3f})")

ax.set_xlabel("ΔC — Incertitude de Complexité", fontsize=13)
ax.set_ylabel("ΔL — Incertitude de Lisibilité", fontsize=13)
ax.set_title("Relation d'Incertitude de Heisenberg-Phi\n"
             f"ΔC · ΔL ≥ ħ_φ/2 = {plancher:.4f}", fontsize=14)
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)
ax.legend(fontsize=10, loc="upper right")
ax.grid(True, alpha=0.3)
ax.set_aspect("equal")
plt.tight_layout()
plt.show()

print(f"ħ_φ = 1/φ = {HBAR_PHI:.6f}")
print(f"Plancher = ħ_φ/2 = {plancher:.6f}")
print(f"Point doré = √(plancher) = {golden_point:.6f}")
```

## 2. Heatmap — Complexité vs Lisibilité

Visualisons la tension quantique pour toutes les combinaisons possibles de ΔC et ΔL.

```python
# Heatmap de la tension quantique
dc_range = np.linspace(0.01, 1.2, 200)
dl_range = np.linspace(0.01, 1.2, 200)
DC, DL = np.meshgrid(dc_range, dl_range)

tension = (DC * DL) / plancher

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.contourf(DC, DL, tension, levels=np.linspace(0, 5, 30), cmap="inferno")
plt.colorbar(im, label="Tension quantique (ΔC·ΔL / plancher)", ax=ax)

# Contour de la frontière quantique (tension = 1)
ax.contour(DC, DL, tension, levels=[1.0], colors=["#FFD700"], linewidths=[3])

# Zones annotées
ax.text(0.15, 0.15, "Super-\ncohérent\n(< 1)", fontsize=11, color="white",
        ha="center", fontweight="bold")
ax.text(0.8, 0.8, "Zone\nclassique\n(> 1)", fontsize=11, color="white",
        ha="center", fontweight="bold")

# Point golden
ax.plot(golden_point, golden_point, "*", color="#FFD700", markersize=20, zorder=5)

ax.set_xlabel("ΔC — Incertitude de Complexité", fontsize=13)
ax.set_ylabel("ΔL — Incertitude de Lisibilité", fontsize=13)
ax.set_title("Heatmap de Tension Quantique Heisenberg-Phi\n"
             "Ligne dorée = frontière quantique (tension = 1)", fontsize=14)
ax.set_aspect("equal")
plt.tight_layout()
plt.show()
```

## 3. Audit en Direct — Tension Quantique

Calculons la tension quantique sur des fichiers réels du projet phi-complexity.

```python
# Audit de fichiers réels
fichiers = [
    ("code_harmonieux.py", "../examples/code_harmonieux.py"),
    ("code_chaotique.py", "../examples/code_chaotique.py"),
]

fig, ax = plt.subplots(figsize=(10, 8))

# Courbe d'incertitude
dc_curve = np.linspace(0.01, 1.5, 300)
dl_curve = plancher / dc_curve
ax.plot(dc_curve, dl_curve, color="#FF6347", linewidth=2, alpha=0.5)
ax.fill_between(dc_curve, 0, np.minimum(dl_curve, 1.5), alpha=0.1, color="red")

# Point golden de référence
ax.plot(golden_point, golden_point, "*", color="#FFD700", markersize=20, zorder=5,
        label="Point Doré (optimum)")

markers = ["o", "s", "D", "^", "v"]
colors = ["#1E90FF", "#FF6347", "#32CD32", "#FF69B4", "#9370DB"]

for idx, (label, path) in enumerate(fichiers):
    try:
        m = auditer(path)
        tension_val = m.get("heisenberg_tension", 0)
        # Recalculer ΔC et ΔL depuis les métriques
        sigma_max_sq = PHI**2 * 100
        h_max = math.log2(PHI**4)
        dc = math.sqrt(m["lilith_variance"] / sigma_max_sq) if m["lilith_variance"] > 0 else 0.01
        dl = min(1.0, m.get("fibonacci_entropy", 0) / h_max) if h_max > 0 else 0.01
        ax.plot(dc, dl, markers[idx % len(markers)], color=colors[idx % len(colors)],
                markersize=12, zorder=5, label=f"{label} (T={tension_val:.3f})")
        print(f"  {label}: ΔC={dc:.4f}, ΔL={dl:.4f}, Tension={tension_val:.4f}, Radiance={m['radiance']}")
    except FileNotFoundError:
        print(f"  {label}: fichier non trouvé")

ax.set_xlabel("ΔC — Incertitude de Complexité", fontsize=13)
ax.set_ylabel("ΔL — Incertitude de Lisibilité", fontsize=13)
ax.set_title("Position Quantique des Fichiers Analysés", fontsize=14)
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_aspect("equal")
plt.tight_layout()
plt.show()
```
