## Objectifs

1. Lire les métriques zéro d'un audit
2. Visualiser la matrice d'interactions
3. Construire le tableau PRE_ZERO / ZERO_CAUSAL / POST_RENAISSANCE
4. Interroger les transitions via `PhiSearch`

```python
from phi_complexity import auditer
from phi_complexity.notebook_helpers import (
    charger_metriques,
    matrice_interactions_zero,
    tableau_zero_morphogenetique,
)
from phi_complexity.search import PhiSearch
```

```python
# Exemple d'audit local
metriques = auditer('../examples/code_harmonieux.py')
for cle in [
    'zero_condition_tension',
    'zero_condition_alignment',
    'zero_clamped_resistance',
    'zero_attractor_convergence',
    'quasicrystal_coherence',
    'quasicrystal_state',
    'zero_morphogenetic_state',
    'zero_morphogenetic_trigger',
    'sync_index',
]:
    print(f'{cle}: {metriques.get(cle)}')
```

```python
# Matrice de correspondance formelle
for ligne in matrice_interactions_zero():
    print(ligne)
```

```python
# Tableau d'interprétation des états
metriques_list = charger_metriques('../examples')
print(tableau_zero_morphogenetique(metriques_list))
```

```python
# Recherche des transitions dans le corpus harvest
search = PhiSearch(workspace_root='..')
resultats = search.chercher_transitions_zero('POST_RENAISSANCE')
print(search.rapport_recherche(resultats, 'Transitions POST_RENAISSANCE'))
```
